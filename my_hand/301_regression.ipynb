```python
import torch 
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python
torch.manual_seed(1)
```

```python
# x data(tensor), shape = (100, 1)
x = torch.unsqueeze(torch.linspace(-1, 1, 100), dim=1)
# noisy y data(tensor), shape = (100, 1)
y = x.pow(2) + 0.2*torch.rand(x.size())

plt.scatter(x.data.numpy(), y.data.numpy())
plt.show()
```

```python
class Net(torch.nn.Module) : 
    def __init__(self, n_feature, n_hidden, n_output): 
        super(Net, self).__init__()
        self.hidden = torch.nn.Linear(n_feature, n_hidden)
        self.predict = torch.nn.Linear(n_hidden, n_output)
        
    def forward(self, x) : 
        x = self.hidden(x)
        x = F.relu(x)
        x = self.predict(x)
        return x
```

```python
net = Net(n_feature=1, n_hidden=10, n_output=1)
print(net.hidden.weight.grad)
print(net.predict.weight.grad)

print(net.hidden.weight)
print(net.predict.weight)
print(net)
```

```python
optimizer = torch.optim.SGD(net.parameters(), lr=.2)
# for regression mean squared loss
loss_func = torch.nn.MSELoss()
```

```python
plt.ion()
```

```python
for t in range(100) : 
    prediction = net(x)

    loss = loss_func(prediction, y)

    # 다음 학습을 위해서 gradient를 비워줍니다
    optimizer.zero_grad() 
    print(net.hidden.weight.grad)
    print(net.hidden.weight.requires_grad)
    # backpropagation을 통해서 gradients를 계산 해 줍니다.
    loss.backward()
    print(net.hidden.weight.grad)
    print(net.hidden.weight.requires_grad)

    # gradient를 적용해서 학습 시켜 줍니다
    optimizer.step()

    if t % 10 == 0 : 
        plt.cla()
        plt.scatter(x.data.numpy(), y.data.numpy())
        plt.plot(x.data.numpy(), prediction.data.numpy(), 'r-', lw=5)
        plt.text(0.5, 0, 'Loss=%.4f' % loss.data.numpy(), fontdict={'size': 20, 'color':  'red'})
        plt.show()
        plt.pause(0.1)
plt.ioff()
```
